`Введите ширину прямоугольника:`

`Площадь прямоугольника: `



```python
a, b = int(input('Введите длинну прямоугольника: ')), int(input('Введите ширину прямоугольника: '))
if a<=0 or b<=0:
  print('Введите положительное число')
else:
  print('Площадь прямоугольника: ', a*b)
```

Задание 5: Напишите программу, которая будет вычислять среднее арифметическое введенных чисел

```python
i = int(input('Введите число чисел '))
spi_sok = []
for j in range(1,i+1):
  a = int(input('Введите ваше число: '))
  spi_sok.append(a)
print(sum(spi_sok)/len(spi_sok))
```

```python
a, b, oper = int(input('Введите первое число ')), int(input('Введите второе число ')), input('Выберите операцию: +, -, *, / ')
if oper == '+': print(a+b)
if oper == '-': print(a-b)
if oper == '*': print(a*b)
if oper == '/': print(a/b)
```
